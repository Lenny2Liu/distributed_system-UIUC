```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import os

with open('big3nodes.txt', 'r') as f:
    lines = f.readlines()

data = [line.strip().split(' ') for line in lines]

# Parse time data as datetime objects

# Group data by transaction name
groups = {}
for row in data:
    name = row[1]
    if name not in groups:
        groups[name] = []
    groups[name].append(row)

# Compute maximum time difference for each transaction
with open('plot_file.txt', 'w') as f:
    for name, group in groups.items():
        initialize_time = min(float(row[0]) for row in group)
        final_time = max(float(row[0]) for row in group)
        time_diff = final_time - initialize_time
        if(time_diff == 0) or (time_diff > 0.005):
            continue
        f.write(f'Transaction {name}: maximum time difference = {time_diff}\n')




```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib.ticker import PercentFormatter

time_differences = []
with open('plot_file.txt', 'r') as f:
    lines = f.readlines()
    time_diff = [line.strip().split('=')[1] for line in lines]
    time_differences = [float(i) for i in time_diff]
sorted_diff = np.sort(time_differences)
n = len(sorted_diff)

ecdf = np.arange(1, n+1) / n

# plot CDF graph
fig, ax = plt.subplots()
ax.plot(sorted_diff, ecdf, linestyle='-', marker='.')
ax.yaxis.set_major_formatter(PercentFormatter(xmax=1))
ax.set_xlabel('Time Differences')
ax.set_ylabel('Empirical CDF')
ax.set_title('CDF of Time Differences')
plt.show()
```
